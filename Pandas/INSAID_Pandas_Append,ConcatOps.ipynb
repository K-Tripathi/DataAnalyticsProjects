```python
import pandas as pd
data_dict = {'id'  :[1,2,3,4,5],
             'name':['Suchit', 'Mohit', 'Mukesh', 'Hemanka', 'Ashish'],
             'language':['English', 'Hindi', 'Punjabi', 'Hindi', 'Hindi']}
df1 = pd.DataFrame(data=data_dict)
df1
```

```python
data_dict2 = {'id' :[6],
              'name' :['Hiren'],
              'language' :['English']}
df2 = pd.DataFrame(data=data_dict2)
df2
```

```python
df3 = df1.append(other=df2, ignore_index=True)
df3
```

```python
data_dict1 = {'id': [1,2,3,4,5],
              'name': ['a', 'b', 'c', 'd', 'e'],
              'sub': ['sub1', 'sub2', 'sub3', 'sub4', 'sub5']
}
data_dict2 = {'id': [1,2,3,4,5],
              'name': ['b', 'c', 'd', 'e', 'f'],
              'sub': ['sub3', 'sub4', 'sub5', 'sub6', 'sub7']
}
df1 = pd.DataFrame(data=data_dict1)
df2 = pd.DataFrame(data=data_dict2)
df_apnd = df1.append(df2)
df_apnd
```

```python
df_concat = pd.concat(objs=[df1,df2], axis=0, ignore_index=True)
df_concat
```

```python
df_concaty = pd.concat(objs=[df1,df2], axis=1, ignore_index=True)
df_concaty
```
