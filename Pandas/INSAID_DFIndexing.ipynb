```python
tips=sns.load_dataset(name='tips', cache=True)
tips.head(10)
```

```python
tips.tail()
```

```python
tips.info()
```

```python
tips.groupby(by=['time']).size()
```

```python
tips.groupby(by=['time']).mean()
```

```python
tips.groupby(by=['time']).agg({'total_bill':'mean'})
```

```python
tips.groupby(by=['time']).count()
```

```python
tips.groupby(by=['time']).agg({'total_bill': 'mean'}).reset_index()
```

```python
tips.groupby(by=['time']).agg({'total_bill': ['min', 'max'],
                              'tip': ['min', 'sum', 'max', 'count']})
```

```python
people = [['Rick', 60, 'O+'], ['Morty', 10, 'O+'], ['Summer', 45, 'O-'], ['Beth', 18, 'B+']]
people_df = pd.DataFrame(data=people, columns=['Name', 'Score', 'Blood Group'])
people_df
```

```python
print('Unique values in blood grp: ', people_df['Blood Group'].unique())
print('Unique count in blood grp: ', people_df['Blood Group'].nunique())
```

```python
people_df['Blood Group'].value_counts()
```

```python
def times2(x):
    return x*2
people_df['Score Times2'] = people_df['Score'].apply(times2)
people_df
```

```python
people_df.sort_values(by='Score', ascending=True, inplace=True)
people_df
```
